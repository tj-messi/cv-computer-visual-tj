```python
import pyautogui
import time

# 设置自动刷新间隔（单位：秒）
refresh_interval = 30  # 每5秒刷新一次

cnt = 0  # 计数器

# 无限循环，每隔 refresh_interval 秒刷新一次页面
while True:
    pyautogui.hotkey('ctrl', 'r')  # 模拟按下 Ctrl + R 刷新页面
    print("页面已刷新")
    cnt += 1  # 计数器加1
    print("刷新次数：" + str(cnt))
    time.sleep(refresh_interval)  # 等待指定的秒数
```
